The bar plot allows for a clear visual comparison of math scores across different genders (male vs. female) and ethnicities (e.g., group A, group B, etc.). By using different colors for each ethnicity, and grouping bars based on gender, it becomes easy to observe any patterns or disparities in math scores across these demographic factors. This can help identify potential areas of concern or interest, such as gender gaps or differences in performance among various ethnic groups.

```python
#1 bar graph
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv('StudentsPerformance.csv')
plt.figure(figsize=(12, 8))
sns.barplot(x='gender', y='math score', hue='race/ethnicity', data=data, ci=None)
plt.title('Comparison of Math Scores Across Gender and Ethnicity')
plt.xlabel('Gender')
plt.ylabel('Math Score')
plt.show()
```

The pie chart provides a clear representation of the proportion of females and males in the dataset. It enables quick comparison of the relative sizes of the two gender groups, aiding in understanding the gender balance or imbalance within the sample.This visualization is useful for communicating gender representation in a visually engaging manner.

```python
#2 pie chart
plt.figure(figsize=(8, 6))
data['gender'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['green', 'yellow'])
plt.title('Female and Male Distribution')
plt.ylabel('')
plt.show()
```


Box plots provide information about the central tendency and variability of the writing scores within each category of parental education. Box plots are effective for comparing distributions across multiple groups simultaneously. In this case, analysts can compare the writing score distributions across different levels of parental education within a single visualization, making it easier to spot trends or differences.
---



```python
# 3 Box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='parental level of education', y='writing score', data=data, palette='Set3')
plt.xlabel('Parental Level of Education')
plt.ylabel('Writing Score')
plt.title('Writing Scores Distribution by Parental Level of Education')
plt.xticks(rotation=45)
plt.show()
```

Pair plots allow for the simultaneous examination of relationships between multiple variables. Coloring data points by gender enables easy differentiation between male and female individuals in the scatterplots. This facilitates the comparison of relationships between scores within each gender group, helping identify any gender-specific trends or patterns.

```python
#4 PAIR PLOT
import seaborn as sns
sns.pairplot(data, hue='gender', palette='Set1')
plt.suptitle('Pairplot of Scores (Colored by Gender)')
plt.show()
```

Line graphs allow analysts to identify trends or patterns in the count of genders across different levels of parental education. By plotting the counts over the categories of parental education, it becomes easier to observe any increasing, decreasing, or fluctuating trends in gender representation

```python
#5 LINE GRAPH
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.read_csv('StudentsPerformance.csv')
gender_parental_counts = data.groupby(['parental level of education', 'gender']).size().unstack()
plt.figure(figsize=(12, 8))
gender_parental_counts.plot(kind='line', marker='o')
plt.title('Count of Gender Across Parental Level of Education')
plt.xlabel('Parental Level of Education')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend(title='Gender')
plt.tight_layout()
plt.show()
```
